**Compilado final**

```python
import matplotlib.pyplot as plt   # Importa la librería para crear gráficos.
import matplotlib.path as mpath   # Importa la librería para trabajar con rutas en los gráficos.
import numpy as np                # Importa la librería para trabajar con arreglos numéricos.

def cartaPlasticidad(Limite_liquido, Indice_plasticidad):
    # Con los datos del limite liquido y el Indice de plasticidad se grafica la ubicación del suelo en la carta de plasticidad.
    plt.plot(Limite_liquido, Indice_plasticidad, 'ro')   # Grafica el punto (Limite liquido, Indice de plasticidad) en rojo.
    plt.vlines(Limite_liquido, 0, 60, 'k', '--')         # Dibuja una línea vertical en el Limite liquido.
    plt.annotate(' LL ', (Limite_liquido, 55))           # Añade la etiqueta 'LL' cerca de la línea vertical.
    plt.annotate(' IP ', (20, Indice_plasticidad + 2))   # Añade la etiqueta 'IP' cerca del punto graficado.
    plt.hlines(Indice_plasticidad, 0, 100, 'k', '--')    # Dibuja una línea horizontal en el Indice de plasticidad.

    # Se establecen los límites de los ejes x,y.
    plt.xlim(0, 100)  # Establece el límite del eje x de 0 a 100.
    plt.ylim(0, 60)   # Establece el límite del eje y de 0 a 60.

    # Para graficar las líneas de la carta de plasticidad utilizamos las ecuaciones conocidas para la Línea A y la Línea U:
    x = np.array([0, 100])  # Crea un arreglo con los valores 0 y 100.
    LineaA = 0.73 * (x - 20)  # Calcula los valores para la Línea A.
    LineaU = 0.9 * (x - 8)    # Calcula los valores para la Línea U.
    plt.annotate('Linea A', (90, 50), rotation=38)  # Etiqueta de la Línea A.
    plt.annotate('Linea U', (60, 45), rotation=45)  # Etiqueta de la Línea U.

    # Estas líneas grafican las Líneas A y U de forma estética.
    plt.plot(x, LineaA, 'darkblue', label="Linea A")  # Grafica la Línea A.
    plt.plot(x, LineaU, 'k:', label="Linea U")        # Grafica la Línea U.

    # Grafica líneas frontera de la carta de plasticidad donde se encuentran los suelos CL-ML.
    plt.hlines(7, 15.7, 29.5, 'm')  # Dibuja una línea horizontal para la frontera CL-ML.
    plt.hlines(4, 12.4, 25.5, 'm')  # Dibuja otra línea horizontal para la frontera CL-ML.

    # Estas líneas permiten que se muestren en la gráfica las etiquetas de las diferentes zonas.
    region_MH = np.array([[50, 0], [50, 22], [100, 58], [100, 0]])  # Define las coordenadas de la región MH.
    region_ML = np.array([[25.5, 4], [12.4, 4], [8, 0], [20, 0], [50, 0], [50, 22]])  # Define las coordenadas de la región ML.
    region_CH = np.array([[50, 22], [100, 58], [100, 60], [75, 60], [50, 38]])  # Define las coordenadas de la región CH.
    region_CL_ML = np.array([[29.5, 7], [15.7, 7], [12.4, 4], [25.5, 4]])  # Define las coordenadas de la región CL-ML.
    region_CL = np.array([[15.7, 7], [29.5, 7], [50, 22], [50, 38]])  # Define las coordenadas de la región CL.

    path_MH = mpath.Path(region_MH)  # Crea el camino para la región MH.
    path_CH = mpath.Path(region_CH)  # Crea el camino para la región CH.
    path_CL = mpath.Path(region_CL)  # Crea el camino para la región CL.
    path_CL_ML = mpath.Path(region_CL_ML)  # Crea el camino para la región CL-ML.
    path_ML = mpath.Path(region_ML)  # Crea el camino para la región ML.

    point = np.array([Limite_liquido, Indice_plasticidad])  # Crea un punto con el Limite liquido y el Indice de plasticidad.
    if path_MH.contains_point(point):  # Verifica si el punto está en la región MH.
        print('El punto se encuentra en la zona MH')
    elif path_CH.contains_point(point):  # Verifica si el punto está en la región CH.
        print('El punto se encuentra en la zona CH')
    elif path_CL.contains_point(point):  # Verifica si el punto está en la región CL.
        print('El punto se encuentra en la zona CL')
    elif path_CL_ML.contains_point(point):  # Verifica si el punto está en la región CL-ML.
        print('El punto se encuentra en la zona CL-ML')
    elif path_ML.contains_point(point):  # Verifica si el punto está en la región ML.
        print('El punto se encuentra en la zona ML')
    else:  # Si el punto no está en ninguna región.
        print('El punto no se encuentra en la carta de plasticidad')

    plt.annotate('CL-ML', (15, 5))  # Añade la etiqueta 'CL-ML'.
    plt.annotate('MH', (80, 20))    # Añade la etiqueta 'MH'.
    plt.annotate('CL', (30, 15))    # Añade la etiqueta 'CL'.
    plt.annotate('CH', (62, 40))    # Añade la etiqueta 'CH'.
    plt.annotate('ML', (35, 5))     # Añade la etiqueta 'ML'.
    plt.annotate('NO EXISTE', (15, 35))  # Añade la etiqueta 'NO EXISTE'.

    # Divide la gráfica cuando el límite líquido es igual a 50 para diferenciar si el suelo es de plasticidad alta o baja.
    plt.vlines(50, 0, 60, 'g')  # Dibuja una línea vertical en 50.

    # Estas líneas mejoran la estética de la gráfica, hacen que se sombreen las diferentes zonas de la carta de plasticidad.
    # Dentro de las variables de la d a la m se guardan las coordenadas que delimitan cada zona.
    d = [50, 50, 100, 100]       # Coordenadas para sombrear la región MH.
    e = [0, 22, 58, 0]           # Coordenadas para sombrear la región MH.
    plt.fill(d, e, 'pink')       # Sombrea la región MH.
    f = [25.5, 12.4, 8, 20, 50, 50]  # Coordenadas para sombrear la región ML.
    g = [4, 4, 0, 0, 0, 22]      # Coordenadas para sombrear la región ML.
    plt.fill(f, g, 'lightgray')  # Sombrea la región ML.
    h = [50, 100, 100, 75, 50]   # Coordenadas para sombrear la región CH.
    i = [22, 58, 60, 60, 38]     # Coordenadas para sombrear la región CH.
    plt.fill(h, i, 'lightgreen') # Sombrea la región CH.
    j = [29.5, 15.7, 12.4, 25.5] # Coordenadas para sombrear la región CL-ML.
    k = [7, 7, 4, 4]             # Coordenadas para sombrear la región CL-ML.
    plt.fill(j, k, 'm')          # Sombrea la región CL-ML.
    l = [15.7, 29.5, 50, 50]     # Coordenadas para sombrear la región CL.
    m = [7, 7, 22, 38]           # Coordenadas para sombrear la región CL.
    plt.fill(l, m, 'bisque')     # Sombrea la región CL.

    # Se grafica la grilla, el título y las etiquetas de los ejes.
    plt.grid()  # Muestra la grilla en la gráfica.
    plt.title("Carta de plasticidad", fontsize=10)  # Añade el título de la gráfica.
    plt.xlabel("Limite liquido", fontsize=10)       # Añade la etiqueta del eje x.
    plt.ylabel("Indice de plasticidad", fontsize=10)  # Añade la etiqueta del eje y.
    plt.legend()  # Muestra la leyenda de la gráfica.
cartaPlasticidad(80,30)
```
