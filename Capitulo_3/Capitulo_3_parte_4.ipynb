**Compilado final**

```python
def cartaPlasticidad(Limite_liquido, Indice_plasticidad):
  plt.xlim(0, 100)                          # Establece el límite del eje x de 0 a 100.
  plt.ylim(0, 60)                           # Establece el límite del eje y de 0 a 60.
  plt.grid()                                # Muestra la grilla en la gráfica.
  plt.title("Carta de plasticidad")         # Añade el título de la gráfica.
  plt.xlabel("Limite liquido [%]")          # Añade la etiqueta del eje x.
  plt.ylabel("Indice de plasticidad [%]")   # Añade la etiqueta del eje y.

  x=np.array([0,100])                       # Crea un arreglo con los valores 0 y 100.
  LineaA =0.73*(x-20)                       # Calcula los valores para la Línea A.
  LineaU = 0.9*(x-8)                        # Calcula los valores para la Línea U.
  plt.plot(x, LineaA, 'darkblue')           # Grafica la Línea A.
  plt.plot(x, LineaU, 'k:')                 # Grafica la Línea U.
  plt.annotate('Linea A', (90,50),rotation=38)  # Etiqueta de la Línea A.
  plt.annotate('Linea U', (60,45),rotation=45)  # Etiqueta de la Línea U.
  plt.vlines(50,0,38,'g')                   # Dibuja una línea vertical en 50.

  plt.plot(Limite_liquido, Indice_plasticidad, 'ro')  # Grafica el punto (Limite liquido, Indice de plasticidad) en rojo.
  plt.vlines(Limite_liquido,0,60,'k','--')            # Dibuja una línea vertical en el Limite liquido.
  plt.hlines(Indice_plasticidad,0,100,'k','--')       # Dibuja una línea horizontal en el Indice de plasticidad.

  plt.annotate('CL-ML', (15,5))                    # Añade la etiqueta 'CL-ML'.
  plt.annotate('MH', (80,20))                      # Añade la etiqueta 'MH'.
  plt.annotate('CL', (30,15))                      # Añade la etiqueta 'CL'.
  plt.annotate('CH', (62,40))                      # Añade la etiqueta 'CH'.
  plt.annotate('ML', (35,5))                       # Añade la etiqueta 'ML'.
  plt.annotate('NO EXISTE', (15,35))               # Añade la etiqueta 'NO EXISTE'.

  colores = {
    "MH": 'pink',
    "ML": 'lightgray',
    "CH": 'lightgreen',
    "CL-ML": 'm',
    "CL": 'bisque'
  }
  regiones = {
    "MH": [[50, 0], [50, 22], [100, 58], [100, 0]],
    "ML": [[25.5, 4], [12.4, 4], [8, 0], [20, 0], [50, 0], [50, 22]],
    "CH": [[50, 22], [100, 58], [100, 60], [75, 60], [50, 38]],
    "CL-ML": [[29.5, 7], [15.7, 7], [12.4, 4], [25.5, 4]],
    "CL": [[15.7, 7], [29.5, 7], [50, 22], [50, 38]]
  }
  for region, vertices in regiones.items():
    x, y = zip(*vertices)
    plt.fill(x, y, color=colores[region], label=region)

cartaPlasticidad(80,30)
```
