**Tercer paso**

```python
#Variables del Sistema:
tamiz_200 = 2                     #% pasa
tamiz_4 = 60                      #% pasa
if tamiz_200 > 50 :               #criterio finos
  print("Mundo de los finos")
else:                             #criterio gruesos
  print("Mundo de las Gruesos")
  if tamiz_4 > 50:                #criterio Gravas
    if 0<tamiz_200<5:             #Primer Criterio
      print("Ocurre algo A")
    elif 5<tamiz_200<12:          #Segundo Criterio
        print("Ocurre algo B")
    else:                         #Tercer Criterio
      print("Ocurre algo C")
  else:                           #criterio Arena
    print("Arenas")
```
