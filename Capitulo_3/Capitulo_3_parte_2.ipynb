```python
frame = pd.DataFrame({ # se crea un DataFrame denominado frame
    "algo" : ["5","6","7"] #Se crea una columna con determinados valores
})
frame
```

```python
frame["operacion"] = frame["algo"] *5 # Se multiplica escalar 5
frame
```

```python
frame["algo"] = frame["algo"].astype("int") # Se convierte tipo int y se sobrescribe la columna.
frame["operacion2"] = frame["algo"] *5      #Se multiplica escalar 5
frame
```

```python
frame["operacion3"] = frame["algo"] * 10 /5 +8 # Simulación operaciones complejas
frame
```

```python
frame["resultado"]=frame["operacion2"]*frame["operacion3"] # Operación entre columnas
frame
```

```python
frame.loc[2] #Datos de la fila 2
```

```python
frame["resultado"].loc[2] #Se vuelve un escalar
```

```python
frame["resultado2"] = frame["operacion3"] * frame["resultado"].loc[2] # columna por escalar 770
frame
```

```python
frame["operacion3"].shift(-1) #Desplaza la columna "operacion3" una posición hacia arriba.
```

```python
frame["resultado3"]=frame["operacion2"]*frame["operacion3"].shift(1) #Realiza la respectiva operación.
frame
```

```python
frame["resultado4"] = frame["resultado3"].fillna(100) #Reemplaza los valores vacíos con 100
frame
```

```python
frame["resultado5"] = frame["resultado3"].replace(np.nan, 100) #Reemplaza los valores vacíos con 100
frame
```
