# **CRUD - Create - Read - Updated - Delete**

**Create**

```python
data = {
    "nombre" : "sensor_2",
    "coordenada_X": 15,
    "coordenada_Y":  20,
    "rol": "distancia",
    "created_at": datetime.datetime.utcnow(),
    "updated_at": datetime.datetime.utcnow()
}
db.sensores.insert_one(data)
```

```python
data_sensor_1 = {
    "distance": 6000,
    "created_at": datetime.datetime.utcnow(),
    "updated_at": datetime.datetime.utcnow()
}
db.Sensor_1.insert_one(data_sensor_1)
```

**Read**

```python
for result in db.Sensor_1.find():
  print(result)
```

```python
result = db.Sensores.find_one({"nombre":"sensor_1"})
print(result)
```

Hacer lo mismo para lectura

**Updated**

```python
result = db.Sensores.update_one(
    {"nombre":"sensor_1"},
    {'$set':
      {'coordenada_X': 8},
     }
)
result = db.Sensores.find_one({"nombre":"sensor_1"})
print(result)
```

**Deleted**

```python
result = db.Sensores.delete_one({"name": "sensor_2"})
for result in db.Sensores.find():
  print(result)
```
