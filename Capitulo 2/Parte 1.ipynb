```python
paises = {
    "Argentina": "Buenos Aires",
    "Brasil": "Brasilia",
    "Chile": "Santiago",
    "Colombia": "Bogotá",
    "Ecuador": "Quito"
}

for pais, capital in paises.items():
    print(f"La capital de {pais} es {capital}")
```

# **Funciones**

```python
def my_function():
  print("Soy una función")
my_function()
```

```python
def my_function(nombre):
  print(nombre + " Texto extra")
my_function("Carlos")
```

```python
def my_function(nombre, apellido):
  print(nombre + " " + apellido)
my_function("Carlos", "Gómez")
```

```python
def my_function(nombre, apellido,edad):
  print(nombre + " " + apellido," y su edad es ",str(edad))
my_function("Carlos", "Gómez",27)
```

```python
def my_function(nombre, apellido, edad):
  print(nombre + " " + apellido," y su edad es ",str(edad))
my_function(edad = 27, apellido = "Gomez", nombre = "Carlos")
```
